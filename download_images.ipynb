```python
import pandas as pd
from slugify import slugify
from google_images_download import google_images_download
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/belisards/dados/main/deputados_br.csv')
df['name_sluge'] = df['nome'].apply(slugify)
df.head(2)
```

```python
downloader = google_images_download.googleimagesdownload() 
```

```python
for index,row in df.iterrows():
    folder = row['name_sluge']
    nome = unidecode.unidecode(row['nome'])
    term = "deputado " + '"' + nome + '"' 
    details = {
      'limit': 30,
      'keywords': term,
      'format': 'jpg',
      'color_type': 'full-color',
      'type': 'photo',
      'image_directory': folder
      }
    downloader.download(details)  
```
